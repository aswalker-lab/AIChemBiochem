## Loading a dataframe
You can load dataframes from a comma separated value (csv) file. csv files have values separated by commans. Excel can open csv files as spreadsheets. During debugging, if you want to look at your input csv file, it can be helpful to open it in Excel.
<br/>
We will look at the Delaney dataset, a solubility dataset, as an example. The source of this dataset is: Delaney, J. J. Chem. Inf. Comput. Sci. 2004, 44, 1000-1005.

```python
import pandas as pd

df = pd.read_csv("delaney.csv")
#this will let us see the beginning of the table
df.head()
```

## Examples of what you can do with pandas
You can use pandas to identify rows that meet a certain condition or find the minimum or maximum values. You can also combine pandas and rdkit to make mol objects from the SMILES strings in the dataframe.

```python
#indentify rows with log(solubility) > -1
df_high_sol = df[df["measured log(solubility:mol/L)"] > -1]
print(df_high_sol)

#count how many rows remain in the data frame
print("There are " + str(len(df_high_sol)) + " compounds with log(solubility) > -1")
```

```python
#add rdkit mol objects to the dataframe
from rdkit import Chem

df["mol"] = df["SMILES"].apply(Chem.MolFromSmiles)
df
```

```python
#identify the most and least soluble compound
print("most soluble compound")
most_soluble_row = df.loc[df["measured log(solubility:mol/L)"].idxmax()]
print(most_soluble_row)
#get the mol object and visualize it
most_soluble_mol = most_soluble_row["mol"]
most_soluble_img = Chem.Draw.MolsToGridImage([most_soluble_mol])
display(most_soluble_img)


print("least soluble compound")
least_soluble_row = df.loc[df["measured log(solubility:mol/L)"].idxmin()]
print(least_soluble_row)
#get the mol object and visualize it
least_soluble_mol = least_soluble_row["mol"]
least_soluble_img = Chem.Draw.MolsToGridImage([least_soluble_mol])
display(least_soluble_img)
```
