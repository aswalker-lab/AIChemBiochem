## Variables and simple mathematical operations
In programming,variables store data of various types such as integers, floats (used to represent numbers that are not integers), strings (text), booleans (true/false values), and other objects which you can create that have custom properties or that you can import from modules others have created

Try to name your variables things that make sense but are not too long

Python is dynamically typed, which means that you can change a variable's type, e.g. you could first create a variable as a number and then later make it text. However, this is generally considered bad practice since it could lead to errors where you expect a variable to be one type and it turns out to be another.

When dealing with integers (called int in python) or floats, you can use basic math operations as shown below.

You can use the print statement to see the value of a variable. Try running the code below.

```python
#any lines that start with a '#' character is a comment, these lines will not run and are meant to provide information about the code
# it is good practice to add comments to your code so others can understand what it does

#addition
a = 1
b = 2
c = a + b

#to see the value of a + b, we can print c
print(c)
#if you want to write this operation as text, you can write a statement where you convert the numbers to strings, using str()
#if you try this statment without converting your numbers to strings first, you will get an error since an int and a string can't
#be added together
print(str(a) + " + " + str(b) + " = " + str(c))

#subtraction, note how we can overwrite the value of c
c = a-b
print(str(a) + " - " + str(b) + " = " + str(c))

#multiplication
a = 2
b = 3
c = a*b
print(str(a) + " * " + str(b) + " = " + str(c))

#division
a = 2
b = 6
c = a/b #c will automatically be a float since a/b does not result in an integer value
print(str(a) + " / " + str(b) + " = " + str(c))
#we can check the current types of our variables using type()
print("type of a: " + str(type(a)))
print("type of c: " + str(type(c)))

#exponents
a = 2
b = 3
c = a**b
print(str(a) +"^" + str(b) + " = " + str(c))

#mod - this returns the remainder from division, which can come in handy sometimes
a = 9
b = 4
c = a%b
print("for a non-divisible number")
print(str(a) + " MOD " + str(b) + " = " +str(c))
print("MOD by divisible number will return zero ")
a = 8
c = a%b
print(str(a) + " MOD " + str(b) + " = " +str(c))

#if you would like to get more familiar with this you can edit some of the variables and see what happens!
```

## Strings

In this cell we will go over some useful operations you can do with strings

```python
# to create a string put the text in quotes
a = "Chemistry is awesome!"

#Here are some useful operations on strings
#lower case the whole string
print(a.lower())
#upper case the whole string
print(a.upper())

#split the string by spaces, this will create a list of all substrings separated by a space
#for more on lists see the next section
print(a.split())
#note that if you put a stirng in quotes in the parenthesis after split you can define a different delimiter than space

#count the number of a certain character or substring in a string
print(a.count("i"))

#find the first or last index of a character or substring
print(a.find(' '))
print(a.rfind(' '))

#get the lenght of a string
print(len(a))

#get a substring of the string, for example we can get the first word and last word using the first and last
#indices of the spaces, we could accomplish the same thing using split

#if you type a[first_index:last_index] you will get all of the text between first_index and last_index, not including the character
#at last_index, note that indices start from zero
#leaving last_index blank will go to the end of the string.
print("first word: " + a[0:a.find(' ')])
print("last word: " + a[a.rfind(' '):])
```

## Data structures: lists and dictionary
Data structures are ways to organize data, many are commonly shared accross different programming languages
In python two of the most useful datastructures are lists and dictionaries

Python also has two other built-in data structures tuples and sets, which can be useful but we will not go over them here. Look them up if you want to learn more.

```python
# to define a list we will use [] 
my_list = []

# if we want to make a list with things already in it you can make it like this
my_list =[1, 2, 3]
#python allows you to create lists of mixed types, but this can cause errors if you try to process a list as if it contains
#only one type so be careful!

#we can add things to lists using append
my_list.append(4)

#we can get items from the list using their index
print("first item in list: " + str(my_list[0]))
print("second item in list: " + str(my_list[1]))

#get the length of a list
print("the lenght of the list is " + str(len(my_list)))

#we can get the maximum or minimum value in a list
print("the maximum value in the list is " + str(max(my_list)))
print("the minimum value in the list is " + str(min(my_list)))

#sum the list
print("the sum of the list is "+ str(sum(my_list)))

#sorting lists
#let's first make a list out of order
my_list = [1, 3, 2, 4]
#the list can then be sorted with, this will change the order of the items in my_list
my_list.sort()
print(my_list)

#sometimes it might be difficult to keep track of what entry is at what index in a list
#it would be easier to look up an item by some other value, such as a string
#this is what dictionaries are for, they have a key which is associated with a value
#an empty dictionary can be created with {}
atomic_mass_dictionary = {}

#to create a dictionary with items already in it, let's use the atomic masses of some elements with their symbol as the lookup
atomic_mass_dictionary = {'H': 1.00784, 'He': 4.002602, 'Li': 6.941}

#we can add additional keys and values to the dictionary, you can also update values for an existing key in the same way
atomic_mass_dictionary['Be'] = 9.012182
atomic_mass_dictionary['B'] = 10.811
print(atomic_mass_dictionary)

#to get a specific value out
print(atomic_mass_dictionary['H'])

#to remove an item from the dictionary, note this will cause an error if the key does not exist in the dictionary
removed_item = atomic_mass_dictionary.pop('H')
print(atomic_mass_dictionary)

#depending on what version of python you are using, dictionaries may be unordered or ordered (after version 3.6) in the order you add keys to them
#but we may want to sort our dictionary in a different way either by key or value

#sort by key
sorted_by_key = sorted(atomic_mass_dictionary.items())
print(sorted_by_key)

#sort by value in increasing order, this will give you a sorted list of tuples (key, value) in increasing order
sorted_by_atomic_mass = sorted(atomic_mass_dictionary.items(), key=lambda item: item[1])
print(sorted_by_atomic_mass)
#we can use this to get the lowest mass
print("element in dictionary with lowest mass is " + sorted_by_atomic_mass[0][0] + " with mass of " + str(sorted_by_atomic_mass[0][1]))

#we can turn this back into a dictionary by casting the list to a dictionary, note that order will only be retained after
#turning it back to a dictionary in more recent versions of python, since you might run code someone will run in an older version
#it is best to just deal with the sorted data in list format
sorted_by_atomic_mass_dic = dict(sorted_by_atomic_mass)
print(sorted_by_atomic_mass_dic)

#to do decreasing order
sorted_by_atomic_mass = sorted(atomic_mass_dictionary.items(), key=lambda item: item[1], reverse=True)
print(sorted_by_atomic_mass)

#you can nest dictionaries or use a list as the value to associate more information with a single key
#it may be easier to use custom objects or pandas data tables which will be discussed more in future assignments
#but for example
atom_dictionary = {}
atom_dictionary['H'] = {'name': 'hydrogen','atomic_mass': 1.00784}
atom_dictionary['He'] = {'name': 'helium','atomic_mass': 4.002602}

#then you can get values like this:
print(atom_dictionary['H']['name'])
print(atom_dictionary['H']['atomic_mass'])

#we can check if a key is in a dictionary using the following
print('H' in atom_dictionary)
print('C' in atom_dictionary)
#we can see that H is in our dictionary, but we never added C so it is not there yet
#you can also use this in statements to check if an item is in a list and in combination with
#if statements discussed in the next cell
#replacing in with not in will return the oppposite value

```

## If statements
If statements allow us to evaluate if something is true so that our program can execute different code under different conditions.

If statements can be single if statements and can also contain elif (else if) statements that will execute if a second specified condition is true (but earlier conditions were not true), and else statements that will execute if none of the previous evaluated statements were true.

You can also nest if statements (have one if statement inside another), which is useful for branching logic. However you want to avoid too deep of a branching or your code will be difficult to read!

Note that python uses indentations to indicate what is inside the if statement.
Some other programming languages use other formats such as braces "{"

```python
#we can use if statements to compare if numbers are greater than each other
a = 2
b = 3
c = 4
#try changing these numbers and seeing how it effects the output

#with if only
if a > b:
    print("a is greater")
#note if b is greater there won't be any code executed and there won't be any output

#with if and else
if a > b:
    print("a is greater")
else:
    print("b is greater or equal to a")
    
#with if, elif, and else
if a > b:
    print("a is greater")
elif a == b: #you use a double equal sign to check if two things are equal
    print("a and b are equal")
else:
    print("b is greater")
#note you would use >= and <= for greater than or equal to or less than or equal to

#Nested ifs:
if a < b:
    if a < c:
        print("a is less than b and c")
    else:
        print("a is less than b but not c")
elif a < c:
    print("a is less than c but not b")
else:
    print("a is not less than b or c")

#ifs with multiple statements, note we can accomplish the same logic as the nested ifs above
if a < b and a < c:
    print("a is less than b and c")
elif a < b:
    print("a is less than b but not c")
elif a < c:
    print("a is less than c but not b")
else:
    print("a is not less than b or c")
    
#not in addition to and you can also use or in if statements
```

## Loops
loops allow you to automatically go through items in a list or dictionary or to repeat actions a specific  number of times
therefore, loops are essential for automating processes
Python has different types of loops, while and for
Loops can also be nested

```python
#while loops continue to loop until a certain condition becomes false, these can be dangerous since if you mess up they
#could become an infinite loop

#for example, lets use a while loop to increase a number until it becomes divisible by 5 but is not 5 itself, starting from 1
i = 1
while (i == 5 or i%5 != 0):
    i += 1 #note this is equalivalent to i = i+1 and will incease the value of i by 1

print("the first number, starting from 1, that is divisible by 5 but that is not 5 is " + str(i))

#this is evaluated like an if statment described above, when the logic statement after while is false the loop will end
#note that != means not equals, so the loop in this case will continue as long as i equals 5 or i mod 5 is not zero

#for loops can be used to iterate through every number in a given range, for example:
for i in range(5):
    print(i)
#will start from 0 and go up to the number before the one in parenthesis

#to start from something other than zero
for i in range(1, 5):
    print(i)

#to move at an increment other than 1 (in this example 2)
for i in range(0, 5, 2):
    print(i)

#you can include logic in the for loop
#for example let's classify all numbers between 0 and 6, not including 6 as even or odd, using %
for i in range(1, 6):
    if i%2 == 0:
        print(str(i) + " is even")
    else:
        print(str(i) + " is odd")
        
#you can also use the for loop to iterate through a list, note this also works for characters in strings
element_list = ['H', 'He', 'Li', 'Be', 'B']
for element in element_list:
    print(element)

#the same thing can be done for dictionaries either by key
atomic_mass_dictionary = {'H': 1.00784, 'He': 4.002602, 'Li': 6.941}
#by key
for element in atomic_mass_dictionary:
    print("element " + element + " has an atomic mass of " + str(atomic_mass_dictionary[element]))
    

#you can nest both for and while loops, although it is best to not do this too many times to avoid readability issues
#for example if we have a list of lists and we want to sum the numbers in each interior list
list_of_lists = [[1], [1, 2], [1,2,3]]
list_of_totals = []
for l in list_of_lists:
    total = 0
    for value in l:
        total += value
    list_of_totals.append(total)
print(list_of_totals)
#note we could avoid needing the inner loop by using the sum function described above
    
#we could also create the list of lists above using nested loops
list_of_lists = []
for i in range(1, 4):
    inner_list = []
    for j in range(1, i+1):
        inner_list.append(j)
    list_of_lists.append(inner_list)
print(list_of_lists)

#note that in using an index that increases each loop, i is usually used (short for index)
#in a nested loop with multiple index variables you can progress through the alphabet from i for each index, j, k, l, etc

#in python we can use for loops to create lists in single lines, which can make code shorter
#this is a stylistic choice

#for example let's say we have a list and and we want to make a second list that contains the items of the first
#all multiplied by two
starting_list = [1, 2, 3, 4,]

#we can either do
multiply_by_two_list = [i*2 for i in starting_list]
print(multiply_by_two_list)

#which is equivalent to
multiply_by_two_list = []
for i in starting_list:
    multiply_by_two_list.append(i*2)
print(multiply_by_two_list)

```

## Modules
You can import libraries to perform more advanced operations
Python comes with standard built-in libraries, you can find a list here:https://docs.python.org/3/library/index.html

Below we will look at some examples using the math, statistics, and random modules
There are also many modules you can install

```python
#to import a module type import an then its name, if it is not a built in module you will have to install it to your environment
#we will go over environments in a future lecture
import math
import statistics

# if you want give your library a shorter name for when you refer to it in code later you can also import it as:
import random as rnd

#the math library gives us methods such as rounding up or down
#round up
print("2.5 rounded up is " + str(math.ceil(2.5)))

#round down
print("2.5 rounded down is " + str(math.floor(2.5)))

#math also contains other useful methods like cos, sin, square root, etc. you can look these up as you need them


#random allows you to get a pseudorandom number between 0 and 1, you can then scale this value to fit in a different range
#there are also other methods that allow random sampling from a population or random integer generation

#between zero and 1
print(rnd.random())

#note if you run this multiple times you will get a different answer each time, you can ensure you get the same answer
#if you set a random seed, which can be important in ensuring results using random numbers are consistent from run to run
#we will discuss this more in future lessons


#the statistics library allows you to do things like the the average and standard deviation from a list
#lets combine it with random to see how the number of random numbers we generate affects the mean and standard deviation
#of the values
list_of_5_rnd_nums = []
for i in range(2):
    list_of_5_rnd_nums.append(rnd.random())
    
list_of_1000_rnd_nums = []
for i in range(1000):
    list_of_1000_rnd_nums.append(rnd.random())
    
print("mean of 5 random numbers: " + str(statistics.mean(list_of_5_rnd_nums)))
print("mean of 1000 random numbers: " + str(statistics.mean(list_of_1000_rnd_nums)))

print("standard deviation of 5 random numbers: " + str(statistics.stdev(list_of_5_rnd_nums)))
print("standard deviation of 1000 random numbers: " + str(statistics.stdev(list_of_1000_rnd_nums)))

#intuatively, the mean of a uniform distribution between 0 and 1, which is what we are sampling from is 0.5
#less intuatively, but thankfully available on wikipedia is that the standard deviation of this uniform distribution
#should be 1 divided by the square root of 12 (https://en.wikipedia.org/wiki/Continuous_uniform_distribution#Properties)
#which is approximately 0.29

#as we sample more, our mean and standard deviation should get closer to the expected values. Try running this multiple
#times to confirm that the 1000 random numbers is usually closer to the expected values than the 5 random numbers

#the statistics library can be used for other statistical calculations as well
#there is also the scipy library, which needs to be installed, that also has various statistical methods
#we will be using more of these in future activities
```
